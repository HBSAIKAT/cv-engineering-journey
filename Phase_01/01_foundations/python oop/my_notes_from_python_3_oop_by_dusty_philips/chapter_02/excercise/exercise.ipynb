## Section 3: Chapter 2 Core Implementation

### ImageHandler Class (Topic 1: Classes)
The `ImageHandler` class demonstrates:
- `__init__()` for initialization
- Instance attributes with underscore naming convention
- Methods that operate on instance data

````python
print("=" * 60)
print("1. ImageHandler Class Demonstration")
print("=" * 60)

# Create an ImageHandler instance
image = ImageHandler("sample.jpg")

print(f"Created ImageHandler with file_path: {image.file_path}")

# Load image (simulated)
image.load()
print(f"Image dimensions: {image.get_dimensions()}")

# Access image data through public method (not direct _data access)
data = image.get_data()
print(f"Image data type: {type(data).__name__}")
print(f"Image data: {data}")

print("\n✓ Chapter 2 Concept: Using methods for controlled access to internal attributes (_data, _width, _height)")
````

### Filter Classes (Topic 1: Inheritance Basics)
The `Filter` base class and concrete filters demonstrate:
- Abstract base class pattern
- `__init__()` methods storing configuration
- Methods that are overridden by subclasses
- Access control with configuration dictionaries

````python
print("\n" + "=" * 60)
print("2. Filter Classes Demonstration")
print("=" * 60)

# Create different filter instances
blur_filter = BlurFilter(radius=5)
grayscale_filter = GrayscaleFilter()
resize_filter = ResizeFilter(800, 600)
enhance_filter = EnhanceFilter(level=1.5)

print(f"Created BlurFilter with config: {blur_filter.config}")
print(f"Created GrayscaleFilter with config: {grayscale_filter.config}")
print(f"Created ResizeFilter with config: {resize_filter.config}")
print(f"Created EnhanceFilter with config: {enhance_filter.config}")

# Demonstrate applying filters
print("\nApplying filters to image:")
blur_filter.apply(image)
grayscale_filter.apply(image)
resize_filter.apply(image)
enhance_filter.apply(image)

print("\n✓ Chapter 2 Concept: Each filter is a separate class with its own __init__ and apply() method")
````

### FilterPipeline Class (Topic 1 & 3: List Management)
The `FilterPipeline` class demonstrates:
- Composition: contains a list of Filter objects
- Methods for managing list content (`add_filter`, `clear`, `count`)
- Method chaining pattern (returning `self`)

````python
print("\n" + "=" * 60)
print("3. FilterPipeline Demonstration")
print("=" * 60)

# Create a new image for this demo
demo_image = ImageHandler("demo.jpg")
demo_image.load()

# Create a pipeline and add filters (method chaining)
pipeline = FilterPipeline()
pipeline.add_filter(BlurFilter(radius=3)) \
        .add_filter(GrayscaleFilter()) \
        .add_filter(ResizeFilter(1024, 768))

print(f"Pipeline contains {pipeline.count()} filters")

# Execute the pipeline
print("\nExecuting pipeline:")
pipeline.execute(demo_image)

print("\n✓ Chapter 2 Concept: FilterPipeline manages a list of Filter objects")
print("✓ Method chaining: add_filter() returns self for fluent API")
````

### RecipeManager Class (Topic 2 & 5: Modules & JSON)
The `RecipeManager` class demonstrates:
- File I/O operations
- JSON serialization with standard library
- Deserializing JSON back into objects

````python
print("\n" + "=" * 60)
print("4. RecipeManager Demonstration")
print("=" * 60)

# Create a recipe manager
manager = RecipeManager(recipe_path="./recipes")

# Create a pipeline to save as recipe
pipeline_to_save = FilterPipeline()
pipeline_to_save.add_filter(BlurFilter(radius=7)) \
                .add_filter(EnhanceFilter(level=1.3))

# Save the pipeline as a JSON recipe
print("Saving pipeline as recipe...")
manager.save(pipeline_to_save, "enhancement_recipe")

# Load the recipe back
print("\nLoading recipe from JSON...")
loaded_pipeline = manager.load("enhancement_recipe")

if loaded_pipeline:
    print(f"Loaded pipeline with {loaded_pipeline.count()} filters")
    print("\nApplying loaded recipe to a new image:")
    
    new_image = ImageHandler("new_image.jpg")
    new_image.load()
    loaded_pipeline.execute(new_image)
    
    print("\n✓ Chapter 2 Concept: JSON module used for serialization")
    print("✓ RecipeManager demonstrates file I/O and object reconstruction")
````

## Section 4: Complete Example and Verification

Let's run the complete demo that combines all components:

````python
print("\n" + "=" * 70)
print("COMPLETE WORKFLOW: Image Processing Pipeline")
print("=" * 70)

# 1. Create image
print("\n1️⃣ Creating ImageHandler...")
workflow_image = ImageHandler("workflow_sample.jpg")
workflow_image.load()
print(f"   ✓ Image loaded: {workflow_image.file_path}")
print(f"   ✓ Dimensions: {workflow_image.get_dimensions()}")

# 2. Build pipeline
print("\n2️⃣ Building filter pipeline...")
workflow_pipeline = FilterPipeline()
workflow_pipeline.add_filter(BlurFilter(radius=2)) \
                 .add_filter(GrayscaleFilter()) \
                 .add_filter(ResizeFilter(640, 480)) \
                 .add_filter(EnhanceFilter(level=1.1))
print(f"   ✓ Pipeline ready with {workflow_pipeline.count()} filters")

# 3. Execute pipeline
print("\n3️⃣ Executing pipeline...")
workflow_pipeline.execute(workflow_image)

# 4. Save image
print("\n4️⃣ Saving processed image...")
workflow_image.save("workflow_output.jpg")

# 5. Save as recipe for reuse
print("\n5️⃣ Saving pipeline as reusable recipe...")
recipe_mgr = RecipeManager(recipe_path="./recipes")
recipe_mgr.save(workflow_pipeline, "workflow_recipe")

# 6. Demonstrate recipe reuse
print("\n6️⃣ Loading and reusing recipe...")
new_workflow_image = ImageHandler("new_sample.jpg")
new_workflow_image.load()
reused_pipeline = recipe_mgr.load("workflow_recipe")
if reused_pipeline:
    reused_pipeline.execute(new_workflow_image)
    new_workflow_image.save("new_output.jpg")

print("\n" + "=" * 70)
print("✅ Complete workflow executed successfully!")
print("=" * 70)
````
