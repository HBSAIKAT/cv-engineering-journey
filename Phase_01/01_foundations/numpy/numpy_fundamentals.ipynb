## **What is numpy?**

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

## **Numpy Arrays Vs Python Sequences**

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

### Indexing & Slicing

```python
# indexing - accessing individual elements in the array 

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
print(a1)
print(a2)
print(a3)
```

```python
a1[0] # first element
```

```python
a2
```

```python
a2[1,2] # element in 2nd row, 3rd column
```

```python
a2[1,0]
```

```python
a3
```

```python
a3[1,0,1] # element in 2nd block, 1st row, 2nd column
```

```python
a3[0,1,0] # element in 1st block, 2nd row, 1st column
```

```python
# Slicing - extracting a portion of the array

a1
```

```python
a1[2:5] # elements from index 2 to 4
```

```python
a1[2:8:1] # elements from index 2 to 7 with step of 1
```

```python
a2
```

```python
a2[0,:] # all elements in the first row
# here a2[0,:] means we want all columns in the first row (index 0) of a2
```

```python
a2[:,2] # all elements in the third column
# here a2[:,2] means we want all rows in the third column (index 2) of a2
```

```python
a2[1:,1:3]
# here a2[1:,1:3] means we want all rows from index 1 to the end and columns from index 1 to 2 (3 is exclusive) of a2
```

```python
a2[:] # all elements in a2
```

```python
a2[::] # all elements in a2 (same as a2[:])
```

```python
a2[::2]
#  here a2[::2] means all rows and columns with a step of 2, so we get every 2nd row and every 2nd column of a2
```

```python
a2[::3]
# here a2[::3] means we want every 3rd row and every 3rd column of a2
```

```python
a2[::-3]
```

```python
a2[:,::3]
```

```python
a2[::2,::3]
# here a2[::2,::3] means we want every 2nd row and every 3rd column of a2
```

```python
a2[::2, 1::2]
# here a2[::2, 1::2] means we want every 2nd row and every 2nd column starting from index 1 of a2
```

```python
a2[1::2, ::3]
# here a2[1::2, ::3] means we want every 2nd row starting from index 1 and every 3rd column of a2
```

```python
a2[0:2,1:] 
# here a2[0:2,1:] means we want rows from index 0 to 1 (2 is exclusive) and all columns from index 1 to the end of a2
```

```python
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
a3[1]
```

```python
a3[::2]
```

```python
a3[0, 1, :]
# here a3[0, 1:2] means we want the first block (index 0) and rows from index 1 to 1 (2 is exclusive) of a3
```

```python
a3[1,2,1:2]
# here a3[1,2,1:2] means we want the second block (index 1) and the third row (index 2) and columns from index 1 to 1 (2 is exclusive) of a3
```

```python
a3[2, 1:, 1:]
# here a3[2, 1::, 1:] means we want the third block (index 2) and all rows from index 1 to the end and all columns from index 1 to the end of a3
```

```python
a3[::2, 0, ::2]
# here a3[0::2, 0::3, ::2] means we want every 2nd block starting from index 0, every 3rd row starting from index 0 and every 2nd column of a3
```
