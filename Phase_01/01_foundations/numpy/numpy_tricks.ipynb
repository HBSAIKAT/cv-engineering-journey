```python
# np.where - Return elements chosen from x or y depending on condition.
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.where.html
a = np.array([11,34,22,56,66,77,86,90,100])
a
```

```python
np.where(a<50) # returns the indices of elements less than 50
```

```python
np.where(a>50, 0, a) # replace elements greater than 50 with 0, keep others unchanged
# np.where(condition, true_value, false_value) 
```

```python
np.where(a%2==0, 0, a) 
# replace even numbers with 0, keep odd numbers unchanged
```

```python
# np.correcoef - Return the Pearson product-moment correlation coefficients.
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html
# correlation coefficient is a measure of the linear relationship between two variables. It ranges from -1 to 1, where:
# -1 indicates a perfect negative linear relationship,
# 0 indicates no linear relationship,
# 1 indicates a perfect positive linear relationship.

salary = np.array([30000, 35000, 40000, 45000, 50000])
experience = np.array([1, 4, 3, 2, 4])
np.corrcoef(salary, experience) # returns the correlation coefficient matrix between salary and experience
```

```python
## Set function
  # - np.union1d
  # - np.intersect1d
  # - np.setdiff1d
  # - np.setxor1d
  # - np.in1d

m = np.array([1,2,3,4,5])
n = np.array([4,5,6,7,8])
print(np.union1d(m, n)) # returns the sorted unique elements that are in either m or n
print(np.intersect1d(m, n)) # returns the sorted unique elements that are in both m and n
print(np.setdiff1d(m, n)) # returns the sorted unique elements that are in m but not in n
print(np.setxor1d(m, n)) # returns the sorted unique elements that are in either m or n but not in both
```
