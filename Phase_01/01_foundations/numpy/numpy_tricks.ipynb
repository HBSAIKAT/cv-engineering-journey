```python
# np.sort - Sort an array along a specified axis.
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.sort.html
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
np.sort(a)[::-1] # sort in descending order
```

```python
b= np.random.randint(1, 100, 24).reshape(6,4)
b
```

```python
np.sort(b)
# row wise sorting
```

```python
np.sort(b, axis=0) # sort each column
```

```python
# np.concatenate - Join a sequence of arrays along an existing axis.
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
np.concatenate((c,d), axis=0) # concatenate along rows (vertical stacking) 
```

```python
np.concatenate((c,d), axis=1) # concatenate along columns (horizontal stacking)
```

```python
# np.expand_dims - Expand the shape of an array.   
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.expand_dims.html
a = np.array([1,2,3,4,5])
a
```

```python
np.expand_dims(a, axis=1) 
# adds a new axis at position 1, transforming the shape from (5,) to (5, 1)
```

```python
np.expand_dims(a, axis=0) # add a new axis at the beginning (row vector)
# transforms the shape from (5,) to (1, 5)
```

```python
# np.where - Return elements chosen from x or y depending on condition.
# resources link: https://numpy.org/doc/stable/reference/generated/numpy.where.html
a = np.array([11,34,22,56,66,77,86,90,100])
a
```

```python
np.where(a > 50) # returns the indices of elements greater than 50
```

```python
np.where(a<50) # returns the indices of elements less than 50
```

```python
np.where(a>50, 0, a) # replace elements greater than 50 with 0, keep others unchanged
# np.where(condition, true_value, false_value) 
```

```python
np.where(a%2==0, 0, a) 
# replace even numbers with 0, keep odd numbers unchanged
```
