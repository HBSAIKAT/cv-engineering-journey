### **Numpy array vs Python lists**

```python
# list
a = [i for i in  range(10000000)]
b = [i for i in  range(10000000, 20000000)]
c= []

```

```python
import time
start = time.time()
for i in range(len(a)):
    c.append(a[i] + b[i])
print(time.time() - start)
```

```python
# numpy

import numpy as np

a = np.arange(10000000)
b = np.arange(10000000, 20000000)

start = time.time()
c = a + b

print(time.time() - start)
```

```python
'''
numpy is faster than list because it is 
implemented in C and uses optimized algorithms 
for operations like addition. Additionally,
numpy arrays are stored in contiguous memory
locations, which allows for faster access and
manipulation compared to lists, which are more 
flexible but less efficient for numerical operations.

''' 
```

```python
# memory
a= [i for i in range(10000000)]
import sys

print(sys.getsizeof(a),"Byte")
```

```python
a = np.arange(10000000)
print(sys.getsizeof(a),"Byte") # default is int64, 8 Byte per element
```

```python
a = np.arange(10000000, dtype=np.int32)
print(sys.getsizeof(a),"Byte")
```

```python
a = np.arange(10000000, dtype=np.int8)
print(sys.getsizeof(a),"Byte")
```

### numpy  better than list because of
- 1. Speed
- 2. Memory Efficiency
- 3. Vectorized Operations

# **Advanced Indexing**

```python
# normal indexing and slicing

a = np.arange(12).reshape(3,4)
a
```

```python
a[1,2]
```

```python
a[1:3,1:3]
```

```python
# Fancy indexing
a

a[[0,1,2]] # row 0,1,2
```

```python
a = np.arange(24).reshape(6,4)
a
```

```python
a[[0,2,4,5]] # row 0,2,4,5
```

```python
a[:,[0,2,3]] # all rows, column 0,2,3
```

```python
# Boolean indexing
```

```python
a = np.random.randint(1,100,24).reshape(6,4) 
# randint(low, high, size) -> low is inclusive, high is exclusive 
a
```

```python
# find all numbers greater than 50

a> 50
```

```python
# boolean masking useing the above boolean array to filter the original array
a[a>50]
# here we are using the boolean array to index the original array, which returns only the elements that satisfy the condition (greater than 50 in this case).
```

```python
# find all numbers greater than 50 and are even
a % 2 == 0
```

```python
a[a % 2 == 0] # even numbers 
```

```python
a[(a>50) & (a % 2 == 0)] # numbers greater than 50 and are even

# & --- > bitwise and its use boolean operation on array
```

```python
# find all numbers not divisible by 7

a[a%7 != 0]
```
