```python
'''
Task:
1. Load an image
2. Convert grayscale
3. Display the image
4. Save the image

     User input image location
     image show or save?
     if save file output name will be input name and sucess message 
'''
```

```python
import cv2

print("Hello, Give your image location or file name: ")
image_location = input()
print("Image location entered: {}".format(image_location))
image = cv2.imread(image_location)
if image is not None:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print("Do you want to show the image or save it? (Type 'yes' to show, anything else to save): ")
    show_image = input().lower()
    if show_image == 'yes':
        cv2.imshow('Grayscale Image', gray)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print("Enter the output file name (with extension): ")
        output_file = input()
        cv2.imwrite(output_file, gray)
        print("Image saved successfully!")
else:
    print('Could not read the image. Please check the file path and try again.')
```
